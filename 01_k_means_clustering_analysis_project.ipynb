### <a name='0' > </a> Import of libraries

```python
# Import the necessary libraries
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions


# Print the version of the imported libraries for reference
print(f'Pandas: {pd.__version__}')
```

### <a name ='1'> </a> Data generation

```python
# Import make_blobs datasets
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1000, centers=4, cluster_std=1.2, center_box=(-8.0, 8.0), random_state=42)[0]

# Create DataFrame
df_blobs = pd.DataFrame(data, columns=['x1', 'x2'])
df_blobs.head()
```

### <a name='2'> </a> Visualization created data

```python
# Create a scatter plot to visualize the clustering of the data points
px.scatter(df_blobs, 'x1', 'x2', width=1000, height=500, title='Clustering analysis: 4 clusters', template='ggplot2')
```

### <a name ='3'> </a> K-means algorithm

#### 2 clusters

```python
# Import the KMeans class from the scikit-learn clustering module
from sklearn.cluster import KMeans

kmeans_2_cluster = KMeans(n_clusters=2)
kmeans_2_cluster.fit(data)
```

- wcss

```python
# Get the Within-Cluster Sum o Squares (WCSS) for the KMeans model with 2 clusters
kmeans_2_cluster.inertia_
```

- classification of data points

```python
y_kmeans_2 = kmeans_2_cluster.predict(data)

df_blobs['y_kmeans'] = y_kmeans_2
df_blobs.head()
```

- centroids

```python
# Create a DataFrame for the cluster centroids from the KMeans model with 2 clusters
centroids = pd.DataFrame(data=kmeans_2_cluster.cluster_centers_, columns=['c1', 'c2'])
centroids
```

- visualization of the created model

```python
fig = px.scatter(df_blobs, 'x1', 'x2', 'y_kmeans', width=1000, height=500, title=f'Alogirthm K-means {kmeans_2_cluster.n_clusters} clusters', template='ggplot2')

fig.add_trace(go.Scatter(x=centroids['c1'], y=centroids['c2'], mode='markers',
                         marker={'size': 10, 'line': {'width': 2, 'color': 'red'}}, showlegend=False))

fig.show()
```

- decision boundaries

```python
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(16,10))
plt.title('Decision boundaries')
plot_decision_regions(data, y_kmeans_2, clf=kmeans_2_cluster, legend=2)
plt.scatter(kmeans_2_cluster.cluster_centers_[:, 0], kmeans_2_cluster.cluster_centers_[:, 1], c='black', s=50)
plt.show()
```

#### <a name='4'> </a> Elbow Method - selection of the optimal number of clusters

```python
wcss = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    wcss.append(round(kmeans.inertia_,2))

wcss
```

```python
wcss = pd.DataFrame(wcss, columns=['wcss'])
wcss = wcss.reset_index()
wcss = wcss.rename(columns={'index': 'clusters'})
wcss['clusters'] += 2
wcss
```

```python
px.line(wcss, x='clusters', y='wcss', width=1000, height=500, title='Within-Cluster-Sum of Squared (WCSS)', template='ggplot2')
```

### <a name='5'> </a> Silhouette

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

# Loop over a range of cluster numbers from 2 to 10
for n_clusters in range(2, 11):

    cluster = KMeans(n_clusters=n_clusters)
    labels = cluster.fit_predict(data)

    # Calculate the silhouette score
    score = silhouette_score(data, labels)
    silhouette_scores.append(score)
    print(f'Number of clusters: {n_clusters}, Silhouette Score: {score:.4f}')
```

```python
# Create a DataFrame to store the number of clusters and corresponding silhouette scores
df_scores = pd.DataFrame({
    'Number of Clusters': range(2, 11),
    'Silhouette Score': silhouette_scores
})

df_scores
```

```python
px.line(df_scores, x='Number of Clusters', y='Silhouette Score', markers=True,
              title='Silhouette Score for Agglomerative Clustering', template='ggplot2')
```
